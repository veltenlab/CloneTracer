# Clonal inference P1

In this notebook I run the clonal inference model on P3. This patient has the following alterations:

* IDH2
* NRAS
* KRAS
* Trisomy chr8

There are 2 time point in the data:

* Diagnosis (Day 0)
* Post-treatment (Day 15)

## Import packages

```python
# Import packages
import os
import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
import itertools
import os
import pickle
import collections

import time
from datetime import datetime
from timeit import default_timer as timer
from datetime import timedelta

import pyro
import pyro.distributions as dist
import pyro.poutine as poutine
from pyro.infer import SVI, Trace_ELBO, TraceEnum_ELBO, config_enumerate
from pyro.optim import Adam, AdagradRMSProp
from torch.distributions import constraints
pyro.set_rng_seed(100)

# import helper functions and tree class form source code files (present in the same directory)
from helper_functions import *
```

```python
torch.set_default_tensor_type(torch.cuda.DoubleTensor)
patient = "P1"
```

# Create tree class object

This class contains all functions required to run the model and to generate diagnostic plots. 

```python
input_file = "../data/input_" + patient + ".json"
t = create_tree_class(input_file, name = patient, mult_samp = True, cnv_celltype = False, gpu = True)
```

# Run model


```python
# select initial 2 mutations based on the proportion of cells covered
t.sel_mutation(initial = True)
```

```python
t.names
```

## Initial trees with chr8 & KRAS

Run the model for 1st 4 possible trees 

```python
t.select_tree(num_iter = 500, init = 300)
```

```python
# print ELBO loss for the last 150 iterations
t.print_elbo(num_iter = 500, init = 350, include = [0,1,2,3])
```

```python
# print the change in parameter values over the last iterations for the selected trees
t.print_params(include = t.tree_indices, it = 100)
```

```python
# plot posterior distributions of the inferred latent variables
t.plot_posterior("all", 0, 1000)
```

## Add IDH2

```python
# add the next mutation IDH2 and select the best tree(s)
t.add_mutation()
t.select_tree(num_iter = 400, init = 300)
```

```python
# print ELBO loss for the last iterations for the selected trees
t.print_elbo(num_iter = 400, init = 300, include = t.tree_indices)
```

```python
# plot posterior distribution of the latent variables for tree 1
t.plot_posterior(variables = "all", tree = 1)
```

## Add NRAS

```python
# add last mutation and select the best tree(s)
t.add_mutation()
t.select_tree(num_iter = 400, init = 300)
```

```python
t.print_elbo(num_iter = 400, init = 300, include = t.tree_indices)
```

```python
t.plot_posterior(variables = "all", tree = 6)
```

# Compute clonal assignments

Here the posterior probability of the clonal assignment for each cells is computed for all selected trees

```python
for i in range(len(t.tree_indices)):
    t.clonal_assignment(tree = i)
```
