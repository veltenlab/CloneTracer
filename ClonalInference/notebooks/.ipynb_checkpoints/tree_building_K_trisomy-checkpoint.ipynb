# Tree building K 

In this notebook I sequentially build the best tree for patient K. The process goes as follows:

1. Build a tree class with the data as attribute
2. Find the 2 most covered mutations and build all potential trees
3. Run SVI on each tree independently
4. Select any tree with lowest ELBO in the last 20 iterations
5. Add the next most covered mutation and generate all potential trees
6. Repeat 3-5 until trees contain all mutations
7. Select one of the best trees and compute clonal assignment probabilities for single-cells

I added RUNX1 and the trisomy in chr8 as additional mutations. It might help find the best tree or fix the clonal assignments...

## Load data

```python
# load data from patient W
with open("data/AKLK_mito_filtered.json") as f:
    data_k = json.load(f)
    
# I will remove RUNX1 and the trisomy in chr8 for the moment
data_k["M"] = np.array(data_k["M"])[:, 0:7]
data_k["N"] = np.array(data_k["N"])[:, 0:7]
data_k["colnames"] = data_k["colnames"][0:7]
    
# add type of mutations: 0 = CNV, 1 = nuclear, 2 = mitochondrial
data_k["type"] = np.array([1, 1, 2, 2, 2, 0])
```

# Build tree K

Here I heuristically build the clonal hierarchy of patient K

## Create tree class 


```python
data = {"M": torch.Tensor(data_k["M"]),
         "N": torch.Tensor(data_k["N"]),
         "mut_type": torch.Tensor([1,1,2,2,2,0]),
         "h_alpha": torch.Tensor([1000.0, 1000.0, 1, 1, 1,2]),
         "h_beta": torch.Tensor([1000.0, 1000.0, 1, 1, 1,100]),
         "af_alpha": torch.Tensor([[44, 43, 1, 1, 1, 3], [0,0,0,0,0,0]]),
         "af_beta": torch.Tensor([[100-44, 100-43, 1, 1, 1, 47], [1,1,1,1,1,1]]),
         "r_cnv": torch.Tensor([0,0,0,0,0,1.5]),
         "names": data_k["colnames"], 
         "barcodes": data_k["cell_barcode"], 
         "umapx": data_k["umapx"], 
         "umapy": data_k["umapy"],
         "class_af": True, 
         "class_assign": torch.cuda.IntTensor(data_k["timepoint"]),
         "class_names": ["day0", "day15"],
         "cnv_celltype": False,                                         
         "celltype": [],
         "celltype_names": [],
         "cnv_ct_mean": [],
         "cnv_ct_sd": []}

k = tree("K", data)
k.sel_mutation(initial = True)
```
