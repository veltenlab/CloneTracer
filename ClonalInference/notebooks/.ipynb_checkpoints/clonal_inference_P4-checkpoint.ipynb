# Clonal inference P4

In this notebook I run the clonal inference model on P4. This patient has the following alterations:


* mt:7075G>A 
* Monosomy in chr7. 


There are 3 time points in the data: 

* Diagnosis (Day 0)
* Remission (Day 21)
* Relapse (Day 105)

In this case the celltype-specific model is used.
